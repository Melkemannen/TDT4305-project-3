```python
# Company A B C and D keywords and budget $$$
global_companies = {
        'A': ["big data", "bloom filters", 3],
        'B': ["flajolet martin", 3],
        'C': ["flajolet martin", "dgim algorithm", 3],
        'D': ["big data", 3],
    }
```

```python
def greedy_algorithm(local_companies, queries):
    # Initial revenue
    revenue = 0
    
    # To-do! update revenue using greedy algorithm
    # for each query
    #  has a bid in []
    #     for each company
    #         if query in company
    #             if company has budget left
    #                 Add to has a bid
    #  if has a bid
    #    Select a random cmpany
    #    Add to revenue, remove from budget
    # return revenue
    for query in queries:
        has_bids = [] #for when there are several bids for a query
        for company, keywords in local_companies.items():
            if query in keywords[:-1] and keywords[-1] > 0:
                has_bids.append(company)
        # Without this the first company is always favored, even if there are others bidding
        # From your answers in ed, i think this is the best solution
        if has_bids:
            selected_company = random.choice(has_bids)
            revenue += 1
            local_companies[selected_company][-1] -= 1
        
    
    return revenue
```

```python
total_revenue = 0
total_trials = 10
print("Starting trials using Greedy Algorithm...")
print("------------------------------------------------")
for i in range(total_trials):
    local_companies = copy.deepcopy(global_companies)
    revenue = greedy_algorithm(local_companies, queries)
    total_revenue = total_revenue + revenue
    print(f"Trial {i+1} - Revenue generated: {revenue}")
print("------------------------------------------------")   
print("Average revenue generated for all trials: ",total_revenue/total_trials)
```

#### 4.2. Balance Algorithm

```python
def balance_algorithm(local_companies, queries):
    # Initial revenue
    revenue = 0
    
    # To-do! update revenue using balance algorithm
    # for each query
    #  has a bid in []
    #     for each company
    #         if query in company
    #             if company has budget left
    #                 Add to has a bid
    #  if has a bid
    #    Select largest budget company
    #    if there are several, choose randomly
    #    Add to revenue, remove from budget
    # return revenue
    for query in queries:
        has_bids = [] #for when there are several bids for a query
        for company, keywords in local_companies.items():
            if query in keywords[:-1] and keywords[-1] > 0:
                has_bids.append(company)
        # Without this the first company is always favored, even if there are others bidding
        # From your answers in ed, i think this is the best solution
        if has_bids:
            max_budget = max([local_companies[company][-1] for company in has_bids]) # in case of ties
            scrooge_mcducks = [company for company in has_bids if local_companies[company][-1] == max_budget] # get the companies with the max budget
            selected_company = random.choice(scrooge_mcducks) # select one randomly
            revenue += 1 # Works if there is only one as well, random item when there is only one item = that one item
            local_companies[selected_company][-1] -= 1
    
    return revenue
```

```python
total_revenue = 0
total_trials = 10
print("Starting trials using Balance Algorithm...")
print("-------------------------------------------")
for i in range(total_trials):
    local_companies = copy.deepcopy(global_companies)
    revenue = balance_algorithm(local_companies, queries)
    total_revenue = total_revenue + revenue
    print(f"Trial {i+1} - Revenue generated: {revenue}")
print("-------------------------------------------")   
print("Average revenue generated for all trials: ",total_revenue/total_trials)
```

#### 5.2. Item-Item Collaborative Filtering

```python
def item_cf(rate_m, tup_mu, neigh):

    
    # To-do! implement a item-item CF using cosine similarity as distance measure
    def mean_of_movie(movie):
        non_zero_ratings = movie[movie > 0]
        return np.mean(non_zero_ratings) if non_zero_ratings.size > 0 else 0

    def adjusted_movie(movie):
        mean = mean_of_movie(movie)
        return np.where(movie > 0, movie - mean, 0)

    # Adjust all movies
    adjusted_movies = np.apply_along_axis(adjusted_movie, axis=1, arr=rate_m)

    # Extract user and movie indices from the query
    movie_index = tup_mu[0] - 1  # Convert to 0-indexed
    user_index = tup_mu[1] - 1  # Convert to 0-indexed

    # Compute cosine similarities
    similarities = cosine_similarity([adjusted_movies[movie_index]], adjusted_movies)[0]
    sorted_indices = np.argsort(similarities)[::-1]  # Sort in descending order
    nearest_indices = [idx for idx in sorted_indices if idx != movie_index][:neigh]  # Exclude the movie itself

    #print("NEAREST MOVIES")
    #print(nearest_indices)
    #print("COSINE SIMILARITY WITH MOVIE", tup_mu[0], "(ADJUSTED)")
    #for idx in nearest_indices:
        #print(f"Movie {tup_mu[0]} vs Movie {idx + 1}: {similarities[idx]}")

    # Calculate prediction for the given user and movie
    numerator = 0
    denominator = 0
    for neighbor_index in nearest_indices:
        neighbor_rating = rate_m[neighbor_index, user_index]  # User's rating for the neighbor movie
        similarity = similarities[neighbor_index]  # Similarity score with the target movie

        if neighbor_rating > 0:  # Only consider movies rated by the user
            numerator += similarity * neighbor_rating
            denominator += similarity
            #print(f"Neighbor {neighbor_index + 1} rating: {neighbor_rating}, similarity: {similarity}")

    #print(f"PREDICTION FOR USER {tup_mu[1]}, MOVIE {tup_mu[0]}")
    if denominator > 0:
        prediction = numerator / denominator
        #print(f"Numerator: {numerator}")
        #print(f"Denominator: {denominator}")
        #print(f"Predicted rating: {prediction}")
    else:
        #print("No valid neighbors to predict rating.")
        prediction = 5050505050505050505050

    return prediction
```
